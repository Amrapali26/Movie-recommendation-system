```python
import ast
```

```python
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
df['genres'] = df['genres'].apply(convert)
```

```python
def converts3(obj):
    L = []
    counter =0
    for i in ast.literal_eval(obj):
        if counter is 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L
```

```python
df['cast']=df['cast'].apply(converts3)
```

```python
def featch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
df['crew'] = df['crew'].apply(featch_director)
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
        
```
